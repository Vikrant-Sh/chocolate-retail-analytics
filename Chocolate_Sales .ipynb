```python
import pandas as pd
import numpy as np
```

```python
sales= pd.read_csv('sales.csv')
products= pd.read_csv('products.csv')
customers= pd.read_csv('customers.csv')
calendar = pd.read_csv('calendar.csv')
stores =pd.read_csv('stores.csv')
```

```python
# Check if the data is in correct format or has any null values using , (.info(),    .isnull().sum()) function
sales.isnull().sum()
```

```python
#-------Sales------
print(sales.isnull().sum())
print(sales.info())
print(sales.duplicated().sum())
```

```python
sales['order_date'] = pd.to_datetime(sales['order_date'])
customers.info()
```

```python
#-----Customers-----
print(customers.isnull().sum())
print(customers.info())
print(customers.duplicated().sum())
```

```python
customers['join_date'] = pd.to_datetime(customers['join_date'])
customers.info()
```

```python
#-----Products------
print(products.isnull().sum())
print(products.info())
print(products.duplicated().sum())
```

```python
#------Stores-----
print(stores.isnull().sum())
print(stores.info())
print(stores.duplicated().sum())
```

```python
#----Calendar----
print(calendar.isnull().sum())
print(calendar.info())
print(calendar.duplicated().sum())
```

```python
calendar['date']= pd.to_datetime(calendar['date'])
calendar.info()
```

```python
# Adding progit margin to the sales table
sales['profit_margin']= round((sales['profit']/sales['revenue'])*100,2)
sales['discount_given']= np.where(sales['discount']>0,'Yes','No')
```

```python
sales.columns
```

```python
#Creating month,wuarter and weekend columns from date column in Calendar table
calendar['month_name']= calendar['date'].dt.month_name()
calendar['quarter']= calendar['date'].dt.quarter
calendar['is_weekend']= np.where(calendar['day_of_week'].isin([5,6]),'Yes','No')
```

```python
calendar.columns
```

```python
 # Creating bins for customrs age group
customers['age_group']= pd.cut(customers['age'],bins=[0,25,35,45,55,100],labels=['18-25','26-35','36-45','46-55','55+'])
```

```python
customers.columns
```

```python
missing_ids = ['P0000', 'P0201']

affected = sales[sales['product_id'].isin(missing_ids)]
print("Total affected rows:", len(affected))
print()
print(affected['product_id'].value_counts())
```

```python
sales_clean =sales[~sales['product_id'].isin(missing_ids)]

print("Original rows :", len(sales))
print("Removed rows  :", len(sales) - len(sales_clean))
print("Clean rows    :", len(sales_clean))
print("Data lost     : 0.98% — safe to proceed")
```
